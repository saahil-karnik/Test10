1) Explain the differences between AI, ML, Deep Learning (DL), and Data
Science (DS).

AI
Definition:
AI is the broad field of creating machines or software that can perform tasks that normally require human intelligence.

Goal:
Enable machines to mimic human decision-making and problem-solving.
AI is the umbrella term covering ML and DL.

ML
Definition:
ML is a subset of AI that focuses on systems that learn from data instead of being explicitly programmed.

Goal:
Build models that improve performance through experience (data).All ML is AI, but not all AI uses ML (e.g., rule-based AI).

DL
Definition:
DL is a subset of Machine Learning that uses neural networks with multiple layers (deep neural networks).

Goal:
Automatically extract complex patterns from large datasets, especially unstructured data.DL is a type of ML but requires much more data and compute power.

DS
Definition:
DS is a broad field that involves extracting insights from data using statistics, programming, and domain expertise.

Goal:
Turn raw data into useful information and decision-support. Data Science may use ML, but it also includes non-ML tasks like data cleaning, visualization, and reporting.

2) What are the types of machine learning? Describe each with one
real-world example.

1. Supervised Learning

What it is:

The model learns from labeled data (input + correct output).
Goal: predict the right output for new unseen data.

How it works:

You give the algorithm examples like:
(Features → Label)

The model finds patterns linking input to output.

Real-world Example:

Email Spam Detection

Input: Email text, sender info

Output (label): Spam or Not spam

The model learns from past labeled emails and predicts if a new email is spam.

2. Unsupervised Learning

What it is:

The model learns from unlabeled data.
Goal: find hidden patterns, groups, or structures.

How it works:

No correct answers given.

Model groups similar items or detects structure in the data.

Real-world Example:

Customer Segmentation

Input: Purchase history, browsing data, demographics

No labels given.

The model groups customers into clusters (e.g., price-sensitive buyers, frequent shoppers).
Used in marketing personalization.

3. Semi-Supervised Learning

What it is:

A mix of labeled and unlabeled data.
Useful when labeling data is expensive but you still have lots of raw data.

Real-world Example:

Medical Image Classification with Limited Annotations

Only a small set of MRI images are labeled by doctors (tumor/no tumor).

Thousands of unlabeled images are also provided.

The model learns patterns from all images, improving accuracy while reducing labeling cost.

4. Reinforcement Learning

What it is:

The model learns by interacting with an environment and receiving rewards or penalties.
Goal: maximize cumulative reward through trial and error.

How it works:

No labeled data.

The agent takes actions and improves strategy over time.

Real-world Example:

Self-Driving Cars

Environment: Real-world road

Actions: Accelerate, brake, steer

Reward: Staying safe, following rules, reaching destination

The car learns optimal decisions based on feedback.

3) Define overfitting, underfitting, and the bias-variance tradeoff in machine
learning.

Overfitting
Definition

Overfitting occurs when a model learns the training data too well, including noise and random fluctuations, causing it to perform poorly on unseen (test) data.

Symptoms

Very high accuracy on training data

Poor generalization to new data

Example

A decision tree that keeps splitting until it perfectly memorizes every training example, but fails on new inputs.

Underfitting
Definition

Underfitting happens when a model is too simple to capture the underlying patterns in the data, resulting in poor performance on both training and test data.

Symptoms

Low accuracy during training

High bias and inability to learn important relationships

Example

Using a straight line (linear model) to fit curved data.

Bias–Variance Tradeoff
Definition

The bias–variance tradeoff explains the balance between two types of errors:

Type of Error	Meaning
Bias	Error from overly simple assumptions → leads to underfitting
Variance	Error from overly complex models sensitive to noise → leads to overfitting



4) What are outliers in a dataset, and list three common techniques for
handling them

Outliers are data points that differ significantly from the majority of observations in a dataset.
They lie far outside the normal range of values and can be caused by:

Measurement or data entry errors

Natural rare events

Sampling anomalies

Impact on models:
Outliers can skew statistical summaries, distort model training, and reduce prediction accuracy.

Three Common Techniques for Handling Outliers
1. Removal (Deleting Outliers)

When used: If outliers are due to errors, noise, or irrelevant extreme cases.

Example Method: Removing values more than 3 standard deviations from the mean.

2. Transformation (Reducing Impact)

When used: Outliers are real but too extreme.

Methods include:

Log transformation

Square root transformation

Box–Cox transformation

Goal: Compress large values so they influence the model less.

3. Capping / Winsorization

What it does: Replace extreme values with the nearest acceptable threshold.

Example:

Replace all values above the 95th percentile with the 95th percentile value.

Replace all values below the 5th percentile similarly.

5) Explain the process of handling missing values and mention one
imputation technique for numerical and one for categorical data.

Handling Missing Values — Process
1. Identify Missing Values

Use methods like .isnull().sum() in Python

Check patterns (random or systematic)

2. Analyze the Reason for Missingness

Three types:

MCAR – Missing Completely at Random

MAR – Missing at Random

MNAR – Missing Not at Random

Understanding this helps select the right strategy.

3. Choose a Handling Method

Common options:

Remove rows/columns with too many missing values

Impute (fill) missing data

Use models that can handle missing data (e.g., XGBoost)

4. Apply the Chosen Technique

Handle numerical and categorical data separately

Ensure no data leakage during train-test splitting

5. Validate Impact

Compare model performance before and after imputation

Ensure the imputation doesn’t distort your data distribution

Imputation Techniques
For Numerical Data: Median Imputation

Replace missing numerical values with the median of that feature.

Median is robust to outliers, unlike mean.

Example:
Missing values in "Age" can be replaced by the median age of the dataset.

For Categorical Data: Mode Imputation

Replace missing categorical values with the most frequent category (mode).

Example:
Missing entries in "Gender" or "Department" can be replaced with the most common category.

```python
#8) Write a Python script to: ● Generate 1000 samples from a normal distribution. ● Introduce 50 random missing values. ● Fill missing values with the column mean. ● Plot a histogram before and after imputation.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)
df = pd.DataFrame({"values": data})

# Introduce 50 random missing values
missing_indices = np.random.choice(df.index, 50, replace=False)
df.loc[missing_indices, "values"] = np.nan

# Save copy before imputation
df_before = df.copy()

# Fill missing values with column mean
df["values"] = df["values"].fillna(df["values"].mean())

# Plot histogram BEFORE imputation
plt.hist(df_before["values"].dropna())
plt.title("Histogram Before Imputation")
plt.show()

# Plot histogram AFTER imputation
plt.hist(df["values"])
plt.title("Histogram After Imputation")
plt.show()
```
